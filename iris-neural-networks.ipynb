```python
import numpy as np
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split, KFold
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=pd.read_csv('iris.csv')
iris.isna()
iris.isnull()
iris
```

```python
iris['SpeciesVals'] = pd.factorize(iris['Species'])[0]
```

```python
iris_x=iris.drop('Id', axis=1)
iris_x=iris_x.drop('Species', axis=1)
iris_x=iris_x.drop('SpeciesVals', axis=1)
iris_y=iris['SpeciesVals']
iris_x
```

```python
iris_x_train, iris_x_test, iris_y_train, iris_y_test = train_test_split(iris_x, iris_y, test_size=0.2, random_state=42)
iris_x_train
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=10)
```

```python
iris_x_array = iris_x_train.values
iris_y_array=iris_y_train.values
iris_x_array
```

```python
sns.scatterplot(data=iris, x='SepalLengthCm', y='SepalWidthCm', hue='Species', palette='Set1')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Length vs Width")
plt.show()
```

```python
sns.scatterplot(data=iris, x='PetalLengthCm', y='PetalWidthCm', hue='Species', palette='Set1')
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Petal Length vs Width")
plt.show()
```

```python
layer_configs = [1, 2]  # Number of hidden layers
node_configs = [4, 6, 8]  # Number of nodes in each layer

# Initialize lists to store the cross-validation results for different architectures
architecture_scores = []


for num_layers in layer_configs:
    for num_nodes in node_configs:
        cv_scores = []  # Scores for this architecture configuration

        for train_idx, val_idx in kf.split(iris_x_array):
            X_train, X_val = iris_x_array[train_idx], iris_x_array[val_idx]
            y_train, y_val = iris_y_array[train_idx], iris_y_array[val_idx]

            # Create the neural network model with the current architecture
            model = tf.keras.Sequential()
            model.add(tf.keras.layers.Input(shape=(4,)))

            for _ in range(num_layers):
                model.add(tf.keras.layers.Dense(num_nodes, activation='relu'))

            model.add(tf.keras.layers.Dense(3, activation='softmax'))

            # Compile the model
            model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

            # Train the model on the training data
            history = model.fit(iris_x_train, iris_y_train, epochs=150, batch_size=20, verbose=0)

            # Evaluate the model on the validation data
            _, accuracy = model.evaluate(X_val, y_val, verbose=0)

            # Append the validation accuracy to the list
            cv_scores.append(accuracy)

        # Calculate the mean accuracy for this architecture configuration
        mean_accuracy = np.mean(cv_scores)

        # Store the results for this architecture configuration
        architecture_scores.append({
            'layers': num_layers,
            'nodes': num_nodes,
            'mean_accuracy': mean_accuracy
        })

# Find the best architecture based on mean accuracy
best_architecture = max(architecture_scores, key=lambda x: x['mean_accuracy'])

print("Best Architecture:")
print(f"Number of Layers: {best_architecture['layers']}")
print(f"Number of Nodes per Layer: {best_architecture['nodes']}")
print(f"Mean Accuracy: {best_architecture['mean_accuracy']:.2f}")
```
